The Collatz Problem
===================

if $a_{n-1}$ is odd, $ a_n = 3 a_{n-1} + 1 $

if $a_{n-1}$ is even, $ a_n = {1\over 2}a_{n-1} $



```python
def collatz(n):
    c = [n]
    while n != 1:
        if n%2==1:
            n = 3*n + 1
        else:
            n = n//2
        c.append(n)
    return(c)
```

```python
collatz(14)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 23
c = collatz(n)
plt.plot(c)
plt.grid(True)
plt.show()
```

It's a little easier to follow the action on a log scale.

```python
plt.semilogy(c)
plt.grid(True)
plt.show()
```

```python
cl = []
for n in range(1,30):
    c = collatz(n)
    cl.append(len(c))
xlocations = np.array(range(len(cl)))+0.5
plt.bar(xlocations,cl)
plt.grid(True)
plt.title('Sequence Length')
plt.show()
```
