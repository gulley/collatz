The Collatz Problem
===================

Take any integer $c_1$ > 0

In general, 

if $c_n$ is odd, compute $ c_{n+1} = 3 c_n + 1 $

and if $c_n$ is even, compute $ c_{n+1} = {1\over 2}c_n $

The [Collatz Conjecture](https://en.wikipedia.org/wiki/Collatz_conjecture) states that this sequence always converges to 1.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Define a function to compute the Collatz sequence.

```python
def collatz(n):
    c = [n]
    while n != 1:
        if n%2==1:
            n = 3*n + 1
        else:
            n = n//2
        c.append(n)
    return(c)
```

Consider the sequence starting at 14.

```python
collatz(14)
```

Here's what it looks like if we plot it.

```python
n = 14
c = collatz(n)
plt.plot(c)
plt.grid(True)
plt.show()
```

It's a little easier to follow the action on a log scale.

```python
plt.semilogy(c)
plt.grid(True)
plt.show()
```

How long does it take for a number starting at n converge?

```python
cl = []
for n in range(1,40):
    c = collatz(n)
    cl.append(len(c))
xlocations = np.array(range(len(cl)))+0.5
plt.bar(xlocations,cl)
plt.grid(True)
plt.title('Sequence Length')
plt.show()
```

So something interesting happens when n=27.

Now let's consider all the sequences that start with the numbers between 2 and 1000.

```python
fig = plt.figure(figsize=(8,5))
for n in range(2,1000):
    c = collatz(n)
    c.reverse();
    ax = plt.semilogy(c,color='#0000FF',linewidth=0.1)
plt.grid(True)
fig
```

Now we can zoom into one specific region.

```python
axlist = fig.axes
ax = axlist[0]
ax.set_xlim((100,120))
fig
```
